```python
import pandas as pd
import numpy as np
```

```python
# A. Extraer la informaci√≥n del archivo.
df = pd.read_csv("/content/country_vaccinations.csv")
df
```

```python
# B. Mostrar la estructura y tipos de datos de cada columna, asegur√°ndote que las columnas con fechas sean del tipo datetime64.
df['date'] = pd.to_datetime(df['date'])
df.info()
```

```python
# c. Determinar la cantidad de vacunas aplicadas de cada compa√±√≠a (con base en c√≥mo lo reporta cada pa√≠s en la columna vaccines, en otras palabras, agrupe por vaccines y realice la sumatoria).
resultado = df.groupby("vaccines")["daily_vaccinations"].sum()
resultado
```

```python
# d. Obtener la cantidad de vacunas aplicadas en todo el mundo.
resultado = int(df['daily_vaccinations'].sum())
resultado
```

```python
# e. Calcular el promedio de vacunas aplicadas por pa√≠s.
country = df.groupby("country")
countries = list(country.groups.keys())
for c in countries:
  sub_df = df[df["country"] == c]
  promedio = sub_df["daily_vaccinations"].mean()
  print(f"Pa√≠s: {c}, Promedio de vacunas diarias: {promedio}")
```

```python
# f. Determinar la cantidad de vacunas aplicadas el d√≠a 29/01/21 en todo el mundo.
vacunas_aplicadas = df[df['date'] == '2021-01-29']
vacunas_aplicadas = vacunas_aplicadas['daily_vaccinations'].sum()
print(f'Cantidad de vacunas aplicadas el d√≠a 29/01/21 en todo el mundo: {int(vacunas_aplicadas)}')
```

```python
# g. Crear un dataframe nuevo denominado conDiferencias que contenga los datos originales y una columna derivada (diferencias) con las diferencias de aplicaci√≥n entre las columnas daily_vaccionations y daily_vaccionations_raw.
new_df = df.copy()
new_df['diferencias'] = new_df['daily_vaccinations'] - new_df['daily_vaccinations_raw']
new_df
```
