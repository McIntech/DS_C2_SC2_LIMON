### A) Mostramos la información

```python
df = pd.read_csv('/content/country_vaccinations.csv')
df
```

### B). Mostrar la estructura y tipos de datos de cada columna para identificar qué operaciones puedes realizar con cada una de ellas, asegurándote que las columnas con fechas sean del tipo datetime64.

```python
df['date'] = pd.to_datetime(df['date'])
df.dtypes
```

### C) Determinar la cantidad de vacunas aplicadas de cada compañía (con base en cómo lo reporta cada país en la columna vaccines, en otras palabras, agrupe por vaccines y realice la sumatoria).

```python
# Forma sin librerías
new_list = []
flat_list_1 = []

for company in df['vaccines']:
  company_strip = []
  company_split = company.split(',')
  for c in company_split:
    company_strip.append(c.strip())
  new_list.append(company_strip)

for sublist in new_list:
  for item in sublist:
    flat_list_1.append(item)



# Forma con pandas (Mas simple)
flat_list_2 = (
    df['vaccines']
      .dropna()
      .str.split(',')
      .explode()
      .str.strip()
      .value_counts()
)

print(flat_list_2)
```

### D. Obtener la cantidad de vacunas aplicadas en todo el mundo.

```python
total_vaccinations = int(
    df['total_vaccinations']
      .dropna()
      .sum()
)

print(f'Total de vacunas aplicadas en todo el mundo: {total_vaccinations}')
```

### E. Calcular el promedio de vacunas aplicadas por país.

```python


promedio_por_pais = (
    df[['country', 'total_vaccinations']]
      .dropna()
      .groupby('country')
      .mean()
      .round(0)
      .astype(int)
)
promedio_por_pais
```

### F. Determinar la cantidad de vacunas aplicadas el día 29/01/21 en todo el mundo.

```python
fecha = '2021-01-29'

result = df[df['date'] == fecha]
total_vac_en_fecha = (
    result['total_vaccinations']
      .dropna()
      .sum()
      .round(0)
      .astype(int)
)
print(f'La cantidad de vacunas aplicada en 29/01/2021 es: {total_vac_en_fecha}')
```

### G. Crear un dataframe nuevo denominado conDiferencias que contenga los datos originales y una columna derivada (diferencias) con las diferencias de aplicación entre las columnas daily_vaccionations y daily_vaccionations_raw.  

```python
conDiferencias = df.copy()
conDiferencias = conDiferencias.fillna(0)

conDiferencias['diferencias'] = (
    conDiferencias['daily_vaccinations'] - conDiferencias['daily_vaccinations_raw']
)

conDiferencias
```

### h. Obtener el periodo de tiempo entre el registro con fecha más reciente y el registro con fecha más antigua.

```python
reciente = df['date'].min()
antigua = df['date'].max()
print(f'La fecha mas reciente es: {reciente} y la mas antigua es: {antigua}')
```

### I. Crear un dataframe nuevo denominado conCantidad que contenga los datos originales y una columna derivada (canVac) con la cantidad de vacunas utilizadas cada día (usar la columna vaccines y separar por el carácter , ).

```python
conCantidad = df.copy()

conCantidad['canVac'] = (
    conCantidad['vaccines']
    .fillna('')
    .apply(lambda x: len([v.strip() for v in x.split(',') if v.strip() != '']))
)

conCantidad
```

### j. Generar un dataframe denominado antes20 con todos los registros que se hayan realizado antes del 20 de diciembre de 2020. **bold text**

```python
antes20 = df.copy()
antes20 = antes20[antes20['date'] < '2020-12-20']
antes20
```

###k. Obtener un dataframe denominado pfizer con todos los registros donde se haya utilizado la vacuna Pfizer.

```python
key_word = 'Pfizer/BioNTech'
Pfizer = df.copy()
Pfizer = Pfizer[Pfizer['vaccines'].fillna('').str.contains(key_word)]
Pfizer
```

### I. Almacenar los dataframes generados (conDiferencias, conCantidad, antes20 y pfizer) en un archivo de Excel denominado resultadosReto.xlsx, donde cada dataframe ocupe una hoja diferente. Se recomienda ver la documentación de pd.ExcelWriter.

```python
import pandas as pd

with pd.ExcelWriter('resultadosReto.xlsx', engine='xlsxwriter') as writer:
    conDiferencias.to_excel(writer, sheet_name='conDiferencias', index=False)
    conCantidad.to_excel(writer, sheet_name='conCantidad', index=False)
    antes20.to_excel(writer, sheet_name='antes20', index=False)
    Pfizer.to_excel(writer, sheet_name='Pfizer', index=False)
```
